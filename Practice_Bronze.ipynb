```
tranFileLoc = '/FileStore/tables/Space_war_transactions_dataset-1.csv'
custFileLoc = '/FileStore/tables/Space_war_customer_dataset.csv'

df = spark.read.format("csv").option("header","true").option("inferSchema","true").load(tranFileLoc)
display(df.head())
```

```
df.write.format("delta").saveAsTable("spaceWar.transaction_bronze")
```

```
customerDf = spark.read.format("csv").option("header","true").option("inferSchema","true").load(custFileLoc)
transactionDf = spark.read.format("csv").option("header","true").option("inferSchema","true").load(tranFileLoc)
```

```
from pyspark.sql.functions import col
customerDf.filter((col("balance") > 15000) & (col("loan") == 'yes')).show(5)
```
